**Linear Regression Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4], dtype=float)
Y = np.array([2.1, 4.3, 6.3, 8.2], dtype=float)
```

```python
m = 0.0   # slope
c = 0.0   # intercept

lr = 0.1     # learning rate (small enough to see smooth convergence)
epochs = 25

loss_history = []
```

```python
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
plt.figure(figsize=(8, 6))

for epoch in range(epochs):
    
    # Prediction
    Y_pred = m * X + c
    
    # Loss
    loss = mse(Y, Y_pred)
    loss_history.append(loss)
    
    # Gradients 
    dm = (-2/len(X)) * np.sum(X * (Y - Y_pred))
    dc = (-2/len(X)) * np.sum(Y - Y_pred)
    
    # Updation rule
    m = m - lr * dm
    c = c - lr * dc
    
    # Clear previous frame
    plt.clf()
    
    # Plot actual data
    plt.scatter(X, Y, s=80, label="Actual Data")
    
    # Plot model prediction line
    x_line = np.linspace(0, 5, 100)
    y_line = m * x_line + c
    plt.plot(x_line, y_line, label=f"Epoch {epoch+1}")
    
    # Draw error lines
    for i in range(len(X)):
        plt.plot([X[i], X[i]], [Y[i], m*X[i] + c], 'r--', alpha=0.5)
    
    plt.title(f"Epoch {epoch+1} | Loss = {loss:.4f} | m = {m:.2f}, c = {c:.2f}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.pause(0.5)

plt.show()
```

```python
#  Loss Curve (Convergence Proof)
plt.figure(figsize=(6, 4))
plt.plot(loss_history, marker='o')
plt.title("Loss vs Epochs")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.grid(True)
plt.show()
```
